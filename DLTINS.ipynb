```python
import pandas as pd
```

```python
dataset = pd.read_csv("DLTINS_20210117_01of01.csv")
```

```python
# stroing the columns to delete from dataset that are containing null values more than 100000
null_columns = [feature for feature in dataset.columns if dataset[feature].isnull().sum() > 100000]
```

```python
len(null_columns)
```

```python
data_set = dataset.drop(null_columns, axis = 1)
```

```python
data_set.shape
```

```python
features_with_null_values = [feature for feature in data_set.columns if data_set[feature].isnull().sum() > 0]
```

```python
len(features_with_null_values)
```

```python
data_set[features_with_null_values].isnull().sum()
```

```python
numerical_features = [feature for feature in features_with_null_values if data_set[feature].dtype != "O"]
numerical_features
```

```python
# to remove the null values from the numerical features
for feature in numerical_features:
    median = data_set[feature].median()
    data_set[feature] = data_set[feature].fillna(median)
```

```python
data_set[numerical_features].isnull().sum()
```

```python
data_set[numerical_features[0]].unique()
```

```python
# removing the null values from the categorical features
categorical_features = [feature for feature in features_with_null_values if data_set[feature].dtype == "O"]
len(categorical_features)
```

```python
for feature in categorical_features:
    data_set[feature] = data_set[feature].fillna("Missing")
```

```python
data_set[categorical_features].isnull().sum()
```

```python
data_set.shape
```

```python
cate_features = [feature for feature in data_set.columns if data_set[feature].dtype == "O"]
len(cate_features)
```

```python
for feature in cate_features:
    labels_ordered = data_set.groupby(feature)[numerical_features[0]].median().sort_values().index
    labels_ordered = {k:i for i, k in enumerate(labels_ordered, 0)}
    data_set[feature] = data_set[feature].map(labels_ordered)
```

```python
data_set.head(10)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 6)
pca.fit(data_set)
data_set = pd.DataFrame(pca.transform(data_set))
data_set.columns = ["FinInstrmGnlAttrbts.Id", "FinInstrmGnlAttrbts.FullNm", "FinInstrmGnlAttrbts.ClssfctnTp", "FinInstrmGnlAttrbts.CmmdtyDerivInd", "FinInstrmGnlAttrbts.NtnlCcy", "Issr"]
data_set.head(10)
```
